# Visualizando distribuições entre categorias Titanic

- Criar uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town").  

- Criar uma visualização que mostre as distribuições do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou não. Que diferença você consegue notar nas distribuições dos passageiros que sobreviveram em relação aos que não sobreviveram?

```python
# importando bibliotecas e estilos.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
# carregando dados direto do seaborn.
titanic = sns.load_dataset('titanic')
titanic.head()
```

**Distribuição do valor da passagem de acordo com a cidade de embarque.**

```python
plt.figure(figsize=(12,5))
plt.title('Valor da passagem por cidade de embarque')
sns.stripplot(x='embark_town', y='fare', data=titanic)
plt.show()
```

- A cidade de Quuenstown apresenta valores mais baixo nas passagens.
- A cidade de Chergourg apresentou valores mais altos, sendo outliers dentre os dados.(motivos não verificado).  
Muito pontos sobre postos não na base de valores não possível detalhar as faixas de valores.

```python
plt.figure(figsize=(12,5))
plt.title('Valor da passagem por cidade de embarque')
sns.swarmplot(x='embark_town', y='fare', data=titanic)
plt.show()
```

- Observa-se que o valor praticado na cidade de Southampton se concentrou abaixo de $100.
- Maiores pagantes observa-se na primeira classe.

**Visualização da distribuição do valor da passagem, cidade, sexo e sobreviventes**

```python
sns.catplot(x='embark_town', y='fare', hue='sex', col='survived', kind='swarm', data=titanic)
plt.show()
```

- Do gráfico de não sobrevintes (lado esquerdo) nota-se a presença marcante do sexo masculino com não sobreviventes.
- Do gráfico de sobreviventes (lado direito) nota-se que o número do sexo feminino é o maior significativamente.

```python
titanic.describe()
```

```python
sns.catplot(x='embark_town', y='fare', hue='sex', col='survived', kind='boxen', data=titanic)
plt.show()
```

- Comparando os gráficos de caixas acima apresentados das distribuições de sexo, as mulheres da segunda classe da cidade Cherbourg tiveram maior indíce de sobreviventes, o mesmo vale para os homens.
- A cidade de Queenstown tiveram menor indíce de sobreviventes.
- O sexo feminio da cidade de Cherbourg e Queenstown tiveram menor indice de não sobreviventes.
- O sexo feminino mostra o maior indíce de sobreviventes entre os que embarcaram na cidades de Southampton.

```python
sns.catplot(x='embark_town', y='fare', hue='sex', col='survived', kind='violin', split=True, data=titanic)
plt.show()
```

- A curva de probabilidade da cidade Southampton mostra que o sexo feminino e masculino é maior dos não-sobreviventes.
- A curva de probabilidade de sobrevivência foi maior na cidade de Cherbourg para ambos os sexos.
