Análise de Dados sobre o consumo de álcool no mundo.

Contexto do exercício:
O consumo excessivo de álcool é um problema grave de saúde pública. Imagine que você é um cientista de dados trabalhando em uma equipe da Organização Mundial da Saúde (OMS), cujo objetivo é desenvolver relatórios sobre o consumo de álcool no mundo.

Foram coletados dados de diferentes países, indicando o consumo de cerveja (beer), licor (spirit) e vinho (wine), e o consumo de litros de álcool por pessoa do país no ano de 2010. Seu trabalho, então, é responder a algumas perguntas sobre os dados que irão ajudar a compor o relatório.

Perguntas a serem respondidas

1. Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?
2. Qual o consumo de licor na França?
3. Quais os 5 países nos quais se consome mais vinho?
4. Quais os países nos quais o consumo de cerveja está acima da média?
5. Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?

```python
# importando biblioteca
import pandas as pd
```

```python
# carregar base de dados
df = pd.read_csv('drinks.txt')
```

```python
# verificar os dados carregados
df.head()
```

```python
# verficar informações estatísticas dos dados carregados
df.describe()
```

```python
# Resposta 1: Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?
print("Média consumo Cerveja = ", df.beer_servings.mean(), ", Desvio padrão = ", df.beer_servings.std())
print("Média consumo Licor = ", df.spirit_servings.mean(), ", Desvio padrão = ", df.spirit_servings.std())
print("Média consumo Vinho = ", df.wine_servings.mean(), ", Desvio padrão = ", df.wine_servings.std())
print("Média consumo Total de álcool = ", df.total_litres_of_pure_alcohol.mean(), ", Desvio padrão = ", df.total_litres_of_pure_alcohol.std())
```

```python
# Observa-se que a cerveja possui maior consumo médio quando comparado com as outras bebidas, 
#sendo expressivo seus números de consumo.
#Observa-se que o desvio padrão para Vinho é muito alto, o que pode indicar grande variação dos dados de consumo do período
#analisado ou valores equivocados.
```

```python
#Resposta 2: Qual o consumo de licor na França? 151
df[df['country']=='France']
```

```python
# Resposta 3 : Quais os 5 países nos quais se consome mais vinho?
df.sort_values(by='wine_servings', ascending = False).head(5)
```

```python
# Resposta 4: Quais os países nos quais o consumo de cerveja está acima da média?
df[df['beer_servings']>101.143103]
```

```python
# Resposta 5: Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?
```
